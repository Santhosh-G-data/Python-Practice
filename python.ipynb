```python
# python cal


"""
Simple Calculator - Python Version
Shows all calculator operations using Python operators
"""

def display_menu():
    """Display calculator menu"""
    print("=" * 40)
    print("SIMPLE CALCULATOR - PYTHON VERSION")
    print("=" * 40)
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")
    print("5. Floor Division (//)")
    print("6. Modulus (%)")
    print("7. Power (**)")
    print("=" * 40)

def calculate(num1, num2, choice):
    """Perform calculation based on user choice"""
    if choice == 1:
        result = num1 + num2
        operation = "+"
    elif choice == 2:
        result = num1 - num2
        operation = "-"
    elif choice == 3:
        result = num1 * num2
        operation = "*"
    elif choice == 4:
        if num2 != 0:
            result = num1 / num2
            operation = "/"
        else:
            return None, "Error: Division by zero!"
    elif choice == 5:
        if num2 != 0:
            result = num1 // num2
            operation = "//"
        else:
            return None, "Error: Division by zero!"
    elif choice == 6:
        if num2 != 0:
            result = num1 % num2
            operation = "%"
        else:
            return None, "Error: Division by zero!"
    elif choice == 7:
        result = num1 ** num2
        operation = "**"
    else:
        return None, "Invalid choice!"
    
    return result, f"{num1} {operation} {num2} = {result}"

def demo_calculations():
    """Show demo calculations with different operators"""
    print("\n" + "=" * 40)
    print("DEMO CALCULATIONS")
    print("=" * 40)
    
    # Using different numbers for demo
    a, b = 15, 4
    
    print(f"Numbers: {a} and {b}")
    print(f"Addition: {a} + {b} = {a + b}")
    print(f"Subtraction: {a} - {b} = {a - b}")
    print(f"Multiplication: {a} * {b} = {a * b}")
    print(f"Division: {a} / {b} = {a / b}")
    print(f"Floor Division: {a} // {b} = {a // b}")
    print(f"Modulus: {a} % {b} = {a % b}")
    print(f"Power: {a} ** {b} = {a ** b}")
    
    # Comparison operators
    print("\n" + "=" * 40)
    print("COMPARISON OPERATORS")
    print("=" * 40)
    print(f"{a} > {b} ? {a > b}")
    print(f"{a} < {b} ? {a < b}")
    print(f"{a} == {b} ? {a == b}")
    print(f"{a} != {b} ? {a != b}")
    print(f"{a} >= {b} ? {a >= b}")
    print(f"{a} <= {b} ? {a <= b}")

def interactive_mode():
    """Interactive calculator mode"""
    while True:
        display_menu()
        
        try:
            choice = int(input("\nEnter your choice (1-7) or 0 to exit: "))
            
            if choice == 0:
                print("Thank you for using Python Calculator!")
                break
            
            if choice < 1 or choice > 7:
                print("Please enter valid choice (1-7)!")
                continue
            
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
            
            result, message = calculate(num1, num2, choice)
            
            if result is None:
                print(f"\n❌ {message}")
            else:
                print(f"\n✅ {message}")
                
                # Show additional info
                print(f"\nResult type: {type(result)}")
                if isinstance(result, float):
                    print(f"Rounded: {round(result, 2)}")
                
        except ValueError:
            print("❌ Please enter valid numbers!")
        except KeyboardInterrupt:
            print("\n\nGoodbye!")
            break

# Main program
if __name__ == "__main__":
    print("\n🐍 PYTHON CALCULATOR PROGRAM")
    print("This demonstrates all operators you've learned!\n")
    
    # Show demo calculations
    demo_calculations()
    
    # Ask if user wants interactive mode
    print("\n" + "=" * 40)
    response = input("Do you want to try interactive mode? (yes/no): ").lower()
    
    if response == 'yes' or response == 'y':
        interactive_mode()
    else:
        print("\nYou can see the calculator web version now!")
        print("Open index.html in your browser!")
```

## MATCH-CASE STATEMENT (Python 3.10+)

What is Match-Case?

It's like a cleaner alternative to multiple if-elif-else statements.


```python
#Basic Syntax:
def get_day_type(day):
    match day.lower():
        case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":
            return "Weekday"
        case "saturday" | "sunday":
            return "Weekend"
        case _:
            return "Invalid day"

print(get_day_type("Monday"))   # Weekday
print(get_day_type("Sunday"))   # Weekend
```

```python
#Match-Case with Conditions:
def check_number(num):
    match num:
        case 0:
            return "Zero"
        case 1 | 2 | 3:
            return "Small number"
        case 4 | 5 | 6:
            return "Medium number"
        case _ if num > 6:
            return "Large number"
        case _:
            return "Negative number"

print(check_number(5))   # Medium number
```

```python
#Match-Case with Patterns:
def process_command(command):
    match command.split():
        case ["hello"]:
            return "Hi there!"
        case ["add", x, y]:
            return f"Sum: {int(x) + int(y)}"
        case ["exit"]:
            return "Goodbye!"
        case _:
            return "Unknown command"

print(process_command("add 5 3"))   # Sum: 8
```

```python
#Practical Example
#Example 1: Word Counter
def word_counter(text):
    """Count frequency of each word in text"""
    words = text.lower().split()
    word_count = {}

    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

    return word_count

sentence = "the cat and the dog and the bird"
result = word_counter(sentence)
print(result) 
```

```python
def is_palindrome(text):
    """Check if a string is a palindrome (ignoring case and spaces)"""
    # Remove spaces and convert to lowercase
    cleaned = text.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

# Test
print(is_palindrome("racecar"))           
print(is_palindrome("A man a plan a canal Panama"))  
print(is_palindrome("hello"))              
```

```python
#Anagram Checker
def are_anagrams(str1, str2):
    """Check if two strings are anagrams"""
    # Remove spaces and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    return sorted(str1) == sorted(str2)

# Alternative using Counter
from collections import Counter

def are_anagrams_counter(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    return Counter(str1) == Counter(str2)

# Test
print(are_anagrams("listen", "silent"))          
print(are_anagrams("hello", "world"))             
print(are_anagrams("Astronomer", "Moon starer"))  
```

```python
#Extract Email Domain
def extract_domain(email):
    """Extract domain name from email address"""
    try:
        username, domain = email.split('@')
        return f"Username: {username}, Domain: {domain}"
    except ValueError:
        return "Invalid email format"

# Test
print(extract_domain("user@example.com"))      
print(extract_domain("invalid-email"))         
print(extract_domain("AJ@gmail.com")) 
```
